```python
folder = os.path.dirname("/content/aclImdb/test/pos/")
folder_list = os.listdir(folder)
print(folder_list)
```

```python
reviews = []
for txt_file in glob.glob(os.path.join(folder, '*.txt')):
  f = open(txt_file, 'r')
  txt = f.read()
  filename = os.path.splitext(os.path.basename(txt_file))[0]
  df_temp = [str(filename), str(txt)]
  #print(df_temp)
  reviews.append(df_temp)
  f.close()
reviews
```

```python
reviews[1]
```

```python
reviews_df = pd.DataFrame(reviews, columns = ['Movie', 'Review'])
```

```python
reviews_df
```

```python
reviews_df[['Movie', 'Rating']] = reviews_df['Movie'].str.split('_', expand = True)
reviews_df
```

```python
f = open('/content/gdrive/MyDrive/Capstone/aclImdb/test/urls_pos.txt')
lines = f.readlines()
lines[1]
```

```python
def movie_link(movie_id):
  link = lines[int(movie_id) - 1]
  link = link.replace('/usercomments\n', '')
  return link
```

```python
reviews_df['Movie'] = reviews_df.Movie.apply(lambda x: movie_link(x))
reviews_df
```
