```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/Capstone/movie_review_emotion.csv', index_col = 0)
df
```

```python
for x in df.columns.values:
  df[x] = df[x].fillna(0)
df
```

```python
df.drop('Review', axis = 1, inplace = True)
df
```

```python
df.columns
```

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```

```python
df = df.groupby('Movie').mean().reset_index()
```

```python
df
```

```python
df = df.melt(id_vars = ["Movie", "Rating"], var_name = "Emotion", value_name = "Strength")
df
```

```python
df_temp = df.loc[df['Emotion'] == 'happiness']
df_temp.sort_values('Strength', ascending = False, inplace = True)
df_temp.head(5)
```

```python
df.sort_values("Movie", inplace = True)
df
```

```python
df_emotions = []
```

```python
print(df.loc[df['Movie'] == 'http://www.imdb.com/title/tt0832971'].sort_values(by = 'Strength',
                                             ascending = False).head(1))
```

```python
for x in df['Movie'].unique():
  entry = df.loc[df['Movie'] == x].sort_values(by = 'Strength',
                                             ascending = False).head(1).values
  print(entry)
  movie = entry[0][0]
  rating = entry[0][1]
  emtotion = entry[0][2]
  df_temp = [movie, rating, emtotion]
  df_emotions.append(df_temp)
```

```python
df_emotions = pd.DataFrame(df_emotions, columns = ['Movie',
                                                   'Rating',
                                                   'Emotion'])
df_emotions
```
